```python
from PyShbak.Processor import General_Processor
```

```python
def text_prepare(input_text:str) -> str:
    out_text = General_Processor.remove_emojis(input_text)
    out_text = General_Processor.remove_hasgtag(out_text)
    out_text = General_Processor.remove_links(out_text)
    out_text = General_Processor.remove_mentions(out_text)
    out_text = General_Processor.remove_punctation(self=0, text = out_text)
    out_text = General_Processor.remove_whitespace(out_text)
    out_text = out_text.lower()
    return out_text
```

```python
data['Processed Text'] = data['text'].apply(text_prepare)
```

```python
test['Processed Text'] = test['text'].apply(text_prepare)
```

```python
test
```

```python
# Preprocess the movie reviews
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stopwords and punctuation
    tokens = [token for token in tokens if token.isalnum() and token not in stop_words]

    # Lemmatize the tokens
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return ' '.join(lemmatized_tokens)
```

```python
# Preprocess movie reviews in the training dataset
data['Processed Text'] = data['Processed Text'].apply(preprocess_text)

# Preprocess movie reviews in the test dataset
test['Processed Text'] = test['Processed Text'].apply(preprocess_text)
```

```python
from PyShbak.Processor import English_Processor
```

```python
English_Processor.english_only()
```

```python
test['Processed Text'] = test['Processed Text'].apply(English_Processor.english_only)
data['Processed Text'] = data['Processed Text'].apply(English_Processor.english_only)
```

```python
# Split the datasets into features and labels
xtrain = data['Processed Text']
ytrain= data['label']
xtest = test['Processed Text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the data from dataset 1
xtrain = vectorizer.fit_transform(xtrain)

# Fit and transform the data from dataset 2
xtest = vectorizer.transform(xtest)
```

```python
import pickle

filename = "tfidf_vectorizer.pkl"
with open(filename, "wb") as file:
    pickle.dump(vectorizer, file)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Create an instance of the classifier
nb_classifier = MultinomialNB()

# Train the classifier
nb_classifier.fit(xtrain, ytrain)
```

```python
y_pred = nb_classifier.predict(xtest)
```

```python
y_pred.shape
```

```python
submission['label'] = y_pred
submission.to_csv('data1.csv',index=False)
```

```python
submission.groupby('label').count().plot(kind='bar')
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
svc = LinearSVC()
param_grid = {
    'C': [0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'loss': ['hinge', 'squared_hinge']
}
```

```python
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, scoring='accuracy', cv=5)
```

```python
grid_search.fit(xtrain, ytrain)
```

```python
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```
